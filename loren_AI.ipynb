# 2. Diabetes Distribution Analysis

## Objective

```python
# ==========================================
# Diabetes Distribution Analysis
# ==========================================

# Count diabetic and non-diabetic patients
diabetes_count = df['diabetes'].value_counts()

print("=" * 50)
print("Diabetes Distribution")
print("=" * 50)

print(diabetes_count)


# ==========================================
# Calculate Percentage Distribution
# ==========================================

diabetes_percentage = (
    df['diabetes']
    .value_counts(normalize=True) * 100
)

print("\n" + "=" * 50)
print("Percentage Distribution")
print("=" * 50)

print(diabetes_percentage)


# ==========================================
# Visualization - Count Plot
# ==========================================

plt.figure(figsize=(8,5))

sns.countplot(
    x='diabetes',
    data=df
)

plt.title("Distribution of Diabetes Patients")
plt.xlabel("Diabetes (0 = No, 1 = Yes)")
plt.ylabel("Patient Count")

plt.show()


# ==========================================
# Pie Chart Visualization
# ==========================================

plt.figure(figsize=(7,7))

df['diabetes'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Diabetes Percentage Distribution")
plt.ylabel("")

plt.show()
```

# 9. HbA1c Level Analysis

## Objective
The objective of this analysis is to understand the distribution of HbA1c levels and examine how HbA1c influences diabetes risk.

## Why This Analysis Matters
HbA1c (Hemoglobin A1c) is one of the most important medical indicators used for diabetes diagnosis.

It represents the average blood sugar level over the past 2–3 months.

Higher HbA1c levels are commonly associated with:
- Poor blood sugar control
- Increased diabetes risk
- Chronic metabolic disorders

Analyzing HbA1c helps identify whether diabetic patients have significantly higher HbA1c levels compared to non-diabetic patients.

## Expected Insights
From this analysis, we expect to:

- Understand HbA1c distribution
- Identify normal and abnormal ranges
- Compare HbA1c levels between diabetic and non-diabetic groups

```python
# ==========================================
# HbA1c Level Analysis
# ==========================================

# Statistical Summary of HbA1c Level
print("=" * 50)
print("HbA1c Level Statistical Summary")
print("=" * 50)

print(df['HbA1c_level'].describe())


# ==========================================
# HbA1c Distribution Histogram
# ==========================================

plt.figure(figsize=(10,5))

sns.histplot(
    df['HbA1c_level'],
    bins=30,
    kde=True
)

plt.title("HbA1c Level Distribution")
plt.xlabel("HbA1c Level")
plt.ylabel("Frequency")

plt.show()


# ==========================================
# HbA1c Boxplot
# ==========================================

plt.figure(figsize=(8,5))

sns.boxplot(
    x=df['HbA1c_level']
)

plt.title("Boxplot of HbA1c Level")

plt.show()


# ==========================================
# HbA1c vs Diabetes Analysis
# ==========================================

plt.figure(figsize=(8,5))

sns.boxplot(
    x='diabetes',
    y='HbA1c_level',
    data=df
)

plt.title("HbA1c Level by Diabetes")
plt.xlabel("Diabetes (0 = No, 1 = Yes)")
plt.ylabel("HbA1c Level")

plt.show()


# ==========================================
# Average HbA1c by Diabetes Status
# ==========================================

print("\n" + "=" * 50)
print("Average HbA1c Level by Diabetes Status")
print("=" * 50)

average_hba1c = df.groupby('diabetes')['HbA1c_level'].mean()

display(average_hba1c)
```

# 21. Outlier Detection Analysis

## Objective
The objective of this analysis is to identify extreme or unusual values (outliers) in numerical healthcare features.

## Why This Analysis Matters
Outliers are values that significantly differ from the normal range of data.

In healthcare datasets, outliers may occur due to:

- Medical abnormalities
- Data entry errors
- Measurement issues
- Rare health conditions

Detecting outliers is important because they can influence:
- Statistical analysis
- Data interpretation
- Machine learning model performance
- Healthcare conclusions

## Expected Insights
From this analysis, we expect to:

- Detect unusual healthcare values
- Understand data spread
- Identify extreme medical cases
- Observe abnormal BMI, glucose, and HbA1c levels

```python
# ==========================================
# Outlier Detection Analysis
# ==========================================

# Select Numerical Columns
numerical_columns = [
    'age',
    'bmi',
    'HbA1c_level',
    'blood_glucose_level'
]


# ==========================================
# Boxplots for Outlier Detection
# ==========================================

plt.figure(figsize=(15,8))

for i, column in enumerate(numerical_columns, 1):
    
    plt.subplot(2, 2, i)
    
    sns.boxplot(
        y=df[column]
    )
    
    plt.title(f'Boxplot of {column}')

plt.tight_layout()

plt.show()


# ==========================================
# IQR Method for Outlier Detection
# ==========================================

print("=" * 60)
print("Outlier Count using IQR Method")
print("=" * 60)

for column in numerical_columns:
    
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[
        (df[column] < lower_bound) |
        (df[column] > upper_bound)
    ]
    
    print(f"{column}: {len(outliers)} outliers")
```

## Interpretation

### Outlier Detection
This analysis identifies extreme values in important healthcare features such as:

- **Age**
- **BMI**
- **HbA1c Level**
- **Blood Glucose Level**

The boxplots visually highlight unusual observations outside the normal range.

### IQR Method
The Interquartile Range (IQR) method is used to statistically detect outliers.

Formula:

- Lower Bound = Q1 − 1.5 × IQR  
- Upper Bound = Q3 + 1.5 × IQR

Values outside this range are considered potential outliers.

- Abnormal HbA1c values

- Blood Glucose Level

- Better model performance
